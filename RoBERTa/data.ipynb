```python
import pandas as pd
from matplotlib import pyplot as plt
```

# Food Reviews Data

```python
# Real Dataset
df = pd.read_csv('./Reviews.csv')
df = df.head(10000)
display(df)
```

```python
#Processed Dataset
processed_df = pd.read_csv('./Roberta_Result.csv')
print(len(processed_df))
```

```python
#Plot Data
display()
df_1 = df[df['Score'] == 1]
df_2 = df[df['Score'] == 2]
df_3 = df[df['Score'] == 3]
df_4 = df[df['Score'] == 4]
df_5 = df[df['Score'] == 5]

label = ["1","2","3","4","5"]
score =[len(df_1), len(df_2),len(df_3),len(df_4),len(df_5)]


def plot_data(label, scores, title):
    plt.figure(figsize=(10,5))
    plt.bar(label, scores,color=['green', 'blue', 'red', 'maroon', 'cyan'])
    plt.title(title)
    plt.show()

plot_data(label, score, "Dataset Scores")


truth_label = ['True Negative', 'True Neutral','True Positive']
nnp_score = [(len(df_1)+len(df_2)), len(df_3), (len(df_4)+len(df_5))]
plot_data(truth_label, nnp_score, "True Scores")
print(f'True Negative:{score[0]+score[1]}\nTrue Neutral:{score[2]}\nTrue Truth:{score[3]+score[4]}')
```
